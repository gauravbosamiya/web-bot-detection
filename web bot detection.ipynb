```python
mm2 = "E:/web_bot_detection_dataset/phase1/data/mouse_movements/humans_and_moderate_bots/"
annotations_path = "E:/web_bot_detection_dataset/phase1/annotations/humans_and_moderate_bots/"

labels = {}

train_file_path = os.path.join(annotations_path, 'train')  

with open(train_file_path, 'r', encoding='utf-8') as train_file:
    for line in train_file:
        parts = line.strip().split()  
        if len(parts) >= 2:
            session_id = parts[0]  
            label = parts[1]      
            labels[session_id] = label  

output_csv = 'output_combined_moderate_with_labels.csv'

with open(output_csv, 'w', newline='', encoding='utf-8') as csv_file:
    writer = csv.writer(csv_file)
    
    header_written = False  

    for subdir in os.listdir(mm2):
        subdir_path = os.path.join(mm2, subdir)
        
        if os.path.isdir(subdir_path):
            for json_file_name in os.listdir(subdir_path):
                json_file_path = os.path.join(subdir_path, json_file_name)
                
                if json_file_name.endswith('.json'): 
                    with open(json_file_path, 'r', encoding='utf-8') as json_file:
                        data = json.load(json_file)
                        
                        session_id = data.get('session_id') 
                        label = labels.get(session_id, 'unknown')  
                        
                        if isinstance(data, dict):  
                            data['label'] = label 
                            if not header_written:
                                writer.writerow(data.keys())  
                                header_written = True
                            writer.writerow(data.values()) 
                        
                        elif isinstance(data, list) and data and isinstance(data[0], dict):  
                            for row in data:
                                row['label'] = label  
                                if not header_written:
                                    writer.writerow(row.keys())
                                    header_written = True
                                writer.writerow(row.values())

print(f"CSV file has been created at {output_csv}.")
```

```python
import numpy as np
import pandas as pd

def extract_time_intervals(times):
    try:
        if pd.isna(times) or not isinstance(times, str):
            return np.array([]) 

        times_list = [int(t.strip()) for t in times.split(",") if t.strip().isdigit()]
        
        if len(times_list) < 2:
            return np.array([]) 

        return np.diff(times_list)  
    except Exception as e:
        print(f"Error processing: {times} - {e}")
        return np.array([])

final_df["time_intervals"] = final_df["mousemove_times"].apply(extract_time_intervals)
```

- total_behaviour: 

   The mouse movement actions of the visitor. These actions can be either move, right click, left click or middle click. The move is written as m(x,y), where the x and y indicate the coordinates of the mouse move. The right, left and middle click are written as c(l), c(r), and c(m) respectively
--------------------------------------------------------------------------------------------------------------------

- mousemove_times:

    the timestamp of when each mouse move was performed in the form of {(t0), (t1), …, (tn)}, where n is the total number of points over which the mouse hovered in each web page.



--------------------------------------------------------------------------------------------------------------------

- mousemove_total_behaviour: 


    the coordinates of the current mouse point in the form of {(x0, y0), (x1, y1), …, (xn, yn)}, where n is the total number of points over which the mouse hovered in each web page.
